```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cohhalf = np.arange(start=0, stop=0.95, step = 0.05)
snr = (2*np.sqrt(cohhalf))/(1-np.sqrt(cohhalf))
cohfull = snr/(1+snr)

plt.subplot(1,2,1)
plt.plot(cohhalf, snr)
plt.xlabel('Coh 1/2')
plt.ylabel('SNR')
plt.subplot(1,2,2)
plt.plot(cohhalf, cohfull)
plt.xlabel('Coh 1/2')
plt.ylabel('Coherence Ceiling')
```
