```python
import numpy as np
#from scipy.special import binom
from scipy.special import genlaguerre
#from scipy.special import hyp1f1
import matplotlib.pyplot as plt
```

```python
# Plot the same as in 
x = np.arange(0, 20.0, 0.01)
fig, ax = plt.subplots()
# ax.set_ylim(-1.0, 1.0)
alpha = 4
order = 5

ax.set_title(f'Generalized Laguerre polynomials upto order {order} for {alpha}')
for xorder in np.arange(0, order):
    y = np.exp(-x)*np.power(x, alpha)*genlaguerre(xorder, alpha)(x)
    ynorm = np.sqrt(np.sum(y*y))
    y = y/ynorm
    ax.plot(x, y, label=f'{xorder}')
plt.legend(loc='best')
plt.show()
```
