```python
import numpy as np
#from scipy.special import binom
from scipy.special import genlaguerre
#from scipy.special import hyp1f1
import matplotlib.pyplot as plt
```

```python
# Plot the same as in 
x = np.arange(0, 12.0, 0.01)
fig, ax = plt.subplots()
ax.set_ylim(-1.0, 1.0)
alpha = 1
order = 6
ax.set_title(f'Generalized Laguerre polynomials upto order {order} for {alpha}')
for xorder in np.arange(0, order):
    ax.plot(x, np.exp(-x)*np.power(x, alpha)*genlaguerre(xorder, alpha)(x), label=f'{xorder}')
plt.legend(loc='best')
plt.show()
```
